```python
import pandas as pd
import numpy as np
from sklearn.metrics import confusion_matrix, roc_curve, RocCurveDisplay
```

```python
df = pd.read_csv("conf_matrix_base.csv")
df = df.to_numpy()
```

```python
y_true = df[:, 1].astype("str")
y_pred = df[:, 0].astype("str")
cm = confusion_matrix(y_true, y_pred, labels=["Phage", "Bacteria"])
cm
```

```python
y_true = (y_true == "Bacteria") * 1
pred_df = pd.read_csv("merged_outputs_versions.csv",usecols = ['score_my'])
pred_df =pred_df.to_numpy()
y_pred = pred_df
```

```python
fpr, tpr, thresholds = roc_curve(y_true=y_true, y_score=y_pred)
```
